```python
import matplotlib.pyplot as plt
import numpy as np
categories = ['lion', 'horse', 'kangaroo']
values1 = [15,24,30]
values2 = [20,18,25]
bar_width = 0.35
bar_positions1 = np.arange(len(categories))
bar_positions2 = bar_positions1 + bar_width
plt.bar(bar_positions1, values1, width=bar_width, label='2021', color='skyblue')
plt.bar(bar_positions2, values2, width=bar_width, label='2022', color='yellow')
plt.xlabel('categories')
plt.plabel('values')
plt.title('grouped bar graph')
plt.legend()
plt.show()
```

```python
import pandas as pd
from pandas import read_csv
path = "/content/drive/MyDrive/rawdata.csv"
df =  pd.read_csv(path)
new_df = df["Duration"].fillna(100);
print(new_df.to_string())
```

```python
x = df["Duration"].mean()
new_df = df["Duration"].fillna(x);
print(new_df.to_string())
y = df["Pulse"].median()
new_df = df["Pulse"].fillna(y);
print(new_df.to_string())
z = df["Maxpulse"].mode()[0]
new_df = df["Maxpulse"].fillna(z);
print(new_df.to_string())
```

```python
df.loc[7, 'Duration'] = 45
print(df.to_string())
```
